```python
from qiskit_aer import Aer
from qiskit import QuantumCircuit, transpile
from qiskit.visualization import plot_state_qsphere, plot_bloch_multivector
```

## Lets writer a helper function to viewing the qsphere

```python
def execute_circuit_qsphere(qc):
    """
    This function returns the circuit diagram, statevector result and plots the qsphere
    """
    sim = Aer.get_backend('aer_simulator_statevector')
    result = sim.run(transpile(qc, sim)).result()

    statevector_res = result.get_statevector(qc)

    circuit_diag = qc.draw()

    q_sphere = plot_state_qsphere(statevector_res)
    return statevector_res, circuit_diag, q_sphere
```

#### Pauli gate `Not X`
This rotates the `|0>` to `|1>` effectively working same as Classical NOT Gate. But this actually rotates `X-Axis` by 180degree

```python
qc = QuantumCircuit(1)
qc.x(0)
qc.save_statevector()
res, img, qsphere = execute_circuit_qsphere(qc)
```

### $R_\phi$ Gate

$\theta$ for horizontal and $\phi$ for vertical axis

so as you can rotate in 3 axis there are 3 gates each.

Where $ -\pi < \theta < \pi$

#### X-Axis
$$
R_X(\theta) = 
\begin{pmatrix}
\cos(\theta) & -i.\sin(\theta) \\
-i.\sin(\theta) & \cos(\theta) 
\end{pmatrix}
$$

#### Y-Axis
$$
R_Y(\theta) = 
\begin{pmatrix}
\cos(\theta) & -\sin(\theta) \\
\sin(\theta) & \cos(\theta) 
\end{pmatrix}
$$

#### Z-Axis
$$
R_Z(\theta) = 
\begin{pmatrix}
e^{-i\theta} & 0 \\
0 & e^{i\theta}
\end{pmatrix}
$$

```python
# Rotate the in Z axis
qc = QuantumCircuit(1)
from math import pi

qc.h(0)
qc.rz(pi/6,0)
qc.save_statevector()
res, img, qsphere = execute_circuit_qsphere(qc)
```

### Universal Gates $U_X$

$U_1$ = 1 degree of freedom
$U_2$ = 2 degree of freedom
$U_3$ = 3 degree of freedom

$$
U_3{(\theta,\phi,\lambda)} =
\begin{pmatrix}
\cos(\theta/2) & -e^{i\lambda}.\sin(\theta/2) \\
e^{i\phi}.\sin(\theta/2) & e^{i{(\phi+\lambda)}}.\cos(\theta/2)
\end{pmatrix}
$$
... Where, $\phi$ for x-axis, $\theta$ for y-axis, $\lambda$ for z-axis

and for this $U_3$
the possible phase as

$0 \leq \theta \leq \pi$

$0 \leq \phi \leq 2.\pi$

$0 \leq \lambda \leq 2.\pi$

### $U_3$

**Deprication Notice**
> u1(theta) = p(theta) = u(0, 0, theta)

> u2(phi, lam) = u(pi/2, phi, lam) = p(pi/2 + phi) sx p(pi/2 lam)

> u3(theta, phi, lam) = u(theta, phi, lam) = p(phi + pi) sx p(theta + pi) sx p(lam)

```python
qc = QuantumCircuit(1)
from math import pi
qc.u(pi/2,pi/2,pi/2,0)
qc.save_statevector()
res, img, qsphere = execute_circuit_qsphere(qc)
```

```python
plot_bloch_multivector(res)
```

### $U_2$

$$
U_2{(\phi,\lambda)} = 1/\sqrt2
\begin{pmatrix}
1 & e^{-i\lambda} \\
e^{i\phi} & e^{i(\phi+\lambda)}
\end{pmatrix}
$$

> $\theta$ is always set as $\pi/2$

```python
qc = QuantumCircuit(1)
from math import pi
qc.u(pi/2,pi/2,pi/2,0)
qc.save_statevector()
res, img, qsphere = execute_circuit_qsphere(qc)
```
