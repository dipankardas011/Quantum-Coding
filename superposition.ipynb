## How to make State Vector
This demonstrates the basic Boch sphere

```python
from qiskit import QuantumCircuit, transpile
from qiskit_aer import Aer

from qiskit.primitives import StatevectorSampler
```

```python
qc = QuantumCircuit(1)
qc.save_statevector()
qc.draw()
```

```python
for i in Aer.backends():
    print(i)
```

```python
backend = Aer.get_backend('aer_simulator_statevector')
```

```python
result = backend.run(transpile(qc, backend)).result()
psi = result.get_statevector(qc)
```

```python
from qiskit.visualization import plot_bloch_multivector
```

```python
plot_bloch_multivector(psi)
```

## Hadamard gate for Creating a superposition

```python
qc_h = QuantumCircuit(1)
qc_h.h(0)
qc_h.save_statevector()
```

```python
qc_h.draw()
```

```python
result_with_h = backend.run(transpile(qc_h, backend)).result()
psi_with_h = result_with_h.get_statevector(qc_h)
```

```python
plot_bloch_multivector(psi_with_h)
```

### Now with Not gate

```python
qc_h_x = QuantumCircuit(1)
qc_h_x.x(0)
qc_h_x.h(0)
qc_h_x.save_statevector()
qc_h_x.draw()
```

```python
psi_with_h_x = backend.run(transpile(qc_h_x, backend)).result().get_statevector(qc_h_x)
```

```python
plot_bloch_multivector(psi_with_h_x)
```
