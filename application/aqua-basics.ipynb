### Uniform Initialization

```python
import numpy as np
from qiskit import QuantumCircuit

# Number of qubits
n = 3
# Uniform statevector: all amplitudes = 1, then normalized
state = np.ones(2**n, dtype=complex) / np.sqrt(2**n)
qc = QuantumCircuit(n)
qc.initialize(state, qc.qubits)
qc.draw('mpl')
```

## Random Initialization

```python
from qiskit.quantum_info import random_statevector

# For 3 qubits, dimension = 2**3 = 8
random_vec = random_statevector(2**3, seed=None)
```

```python
from qiskit import QuantumCircuit
# Using QuantumCircuit.initialize
qc_rand = QuantumCircuit(3)
qc_rand.initialize(random_vec, qc_rand.qubits)

# Visualize
qc_rand.draw('mpl')
```

```python
qc_rand.decompose().draw('mpl')
```
